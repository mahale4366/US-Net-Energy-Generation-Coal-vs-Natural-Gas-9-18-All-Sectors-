```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel (r'C:\Users\matth\Documents\Excel EIA Energy\CREN1.xlsx')
df.index = df.index +9
df.Coal = df.Coal.div(1000000)
df.Total_Renewable_Generation = df.Total_Renewable_Generation.div(1000000)
plt.plot(df.Coal, color='black',label=('Coal'))
plt.plot(df.Total_Renewable_Generation, color='green',label=('Renewable'))
plt.xlabel('Year')
plt.ylabel('Billion Megawatt Hours')
plt.legend()
plt.xlim(9,18)
plt.title('US Net Energy Generation: \n Coal vs Total Renewable Generation 2009-2018 (All Sectors)')

plt.show()
```
