```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import plotly
import cufflinks as cf
cf.go_offline()
import seaborn as sns
```

```python
df = pd.read_excel(r'C:\Users\matth\Documents\DC GHG\2006-2017 DCGHG.xlsx', sheet_name='Source')
df1 = pd.read_excel(r'C:\Users\matth\Documents\DC GHG\2006-2017 DCGHG.xlsx', sheet_name='Type')
df
#Source DOEE 2006-2017 Greenhouse Gas Inventory: https://doee.dc.gov/service/greenhouse-gas-inventories
```

```python
df['Year'].astype(str)
df1['Year'].astype(str)
```

```python
df.set_index('Year',inplace=True)
```

```python
plt.subplot
fig1 = df.plot()
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('DC Mega Tons CO2 Emitted by Source (2009-2017)')
plt.xlabel('Year')
plt.ylabel('MTCO2e')
plt.show()

fig2 = sns.lineplot(x='Year', y='MTCO2e',data=df1,hue='Type', palette='Set1')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('DC Mega Tons CO2 Emitted by Type (2009-2017)')
plt.xlabel('Year')
plt.ylabel('MTCO2e')
plt.show()
```
