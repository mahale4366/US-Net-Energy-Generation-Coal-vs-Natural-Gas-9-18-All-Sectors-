```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel(r'C:\Users\matth\Documents\Excel EIA Energy\CO2 Emissions by state 2005-2017.xlsx')
df.index = df.index + 2003
Ab_percent = df.Alabama / df.US
Ab_percent[2:]
plt.plot(Ab_percent[2:], color='red')
y_pos = np.arange(2005,2016.25)
plt.xticks(y_pos)
plt.xlim(2005,2016)
plt.title('Alabama, Percent of US Total CO2 Emissions \n 2005-2016')
plt.xlabel('Year')
plt.ylabel('Percent CO2 Emissions')
plt.show()
```
