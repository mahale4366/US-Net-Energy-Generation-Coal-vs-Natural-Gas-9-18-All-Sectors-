```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline

df = pd.read_html('https://www.eia.gov/electricity/state/florida/index.php')
df = df[0]
df.set_index('Item',inplace=True)
df = df[:-1]
df.dropna(axis=0, inplace=True)
df.drop(['Rank'],axis=1, inplace=True)

df1 = df.iloc[7,0]
df1 = int(df1)
df2 = df.iloc[8,0]
df2 = int(df2)
df3 = df.iloc[9,0]
df3 = int(df3)

fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.bar(1, df1, .5, color='red', label='Sulfur Dioxide')
ax.bar(2, df2, .5, color='blue', label='Nitrogen Dioxide')
ax.bar(3, df3, .5, color='black', label='CO2 Thousand Metric Tons')
tick = np.arange(1,3.5)
ax.set_xticks(tick)
ax.set_xlim(0.5,3.5)
ax.set_ylim(0, 110000,10000)
ax.set_ylabel('Emissions Short Tons')
x_labels = ['Sulfur Dioxide','Nitrogen Dioxide','CO2 (Thousand \n Metric Tons)']
ax.set_xticklabels(x_labels)
ax.set_title('Florida Energy Production Emissions, 2018')
plt.show()
```
