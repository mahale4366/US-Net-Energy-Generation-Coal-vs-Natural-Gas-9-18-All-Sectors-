```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel (r'C:\Users\matth\Documents\CNG1.xlsx')
df.index = df.index + 9

plt.plot(df.Coal, color='black', label='Coal')
plt.plot(df.Natural_Gas, color='green', label='Natural Gas')
plt.title('US Net Energy Generation Coal vs Natural Gas 9-18 (All Sectors)')
plt.xlabel('Year')
plt.ylabel('Billion Megawatt Hours')
plt.legend()
plt.xlim(9,18)
plt.show()
```
