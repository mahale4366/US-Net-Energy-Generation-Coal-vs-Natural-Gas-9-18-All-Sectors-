```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel(r'C:\Users\matth\Documents\Excel EIA Energy\CNGREN1.xlsx')
df.index = df.index+9
df = df/1000000

plt.plot(df.Coal, color='black',label='Coal')
plt.plot(df.Natural_Gas,color='red', label='Natural Gas')
plt.plot(df.Total_Renewable_Generation, color='green',label='Renewable')
plt.xlabel('Year')
plt.ylabel('Billion Megawatt Hours')
plt.title('US Net Energy Generation: \n Coal vs Natural Gas vs Renewable \n 2009-2018 (All Sectors)')
plt.legend()
plt.xlim(9,18)
plt.show()
```
