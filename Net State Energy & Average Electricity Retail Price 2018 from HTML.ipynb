```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_html('https://www.eia.gov/electricity/state/florida/index.php')
df = df[0]

df1 = pd.read_html('https://www.eia.gov/electricity/state/alabama/index.php')
df1 = df1[0]

df2 = pd.read_html('https://www.eia.gov/electricity/state/georgia/index.php')
df2 = df2[0]

df3 = pd.read_html('https://www.eia.gov/electricity/state/unitedstates/index.php')
df3 = df3[0]

plt.style.use('default')
flnet = int(df.iloc[4,1])
abnet = int(df1.iloc[4,1])
ganet = int(df2.iloc[4,1])

flprice = float(df.iloc[18,1])
abprice = float(df1.iloc[18,1])
gaprice = float(df2.iloc[18,1])
usprice = float(df3.iloc[18,1])

fig1 = plt.figure()
ax1 = fig1.add_axes([0.1,0.1,0.6,0.6])
ax1.bar(.75, flnet, .5, color='blue')
ax1.bar(1.5, abnet, .5, color='red')
ax1.bar(2.25, ganet, .5, color='black')

tick = np.linspace(.75,2.25,num=3)
ax1.set_xticks(tick)
x_labels = ['Florida','Alabama','Georgia']
ax1.set_xticklabels(x_labels)
ax1.set_title('Net Energy Generation by State, 2018')
ax1.set_xlabel('States')
ax1.set_ylabel('10 Million Megawatts')
ax1.grid(which='major', axis='y', linestyle='-',color='black', linewidth=.1)

fig2 = plt.figure()
ax2 = fig2.add_axes([0.1,0.1,0.8,0.6])
ax2.bar(.75, flprice, .5, color='blue')
ax2.bar(1.5, abprice, .5, color='red')
ax2.bar(2.25, gaprice, .5, color='black')
ax2.bar(3, usprice, .5, color='dimgray')

tick2 = np.linspace(.75,3,4)
x_labels2 = ['Florida','Alabama','Georgia','United States']
ax2.set_xlim(0.25,3.5)
ax2.set_xticks(tick2)
ax2.set_xticklabels(x_labels2)
ax2.set_ylim(9,11)
ax2.set_title('Average Retail Price, 2018 (cents/kWh)')
ax2.set_xlabel('States')
ax2.set_ylabel('cents/kWh')
ax2.grid(which='major', axis='y', linestyle='-',color='black', linewidth=.1)

plt.show()
```
