```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel(r'C:\Users\matth\Documents\Excel EIA Energy\New folder\Florida Energy Total 2009-2018.xlsx')
df = df/1000000
y_pos = np.arange(len(df))
plt.bar(1,df.Natural_gas[0], color='red', label='Natural Gas')
plt.bar(2,df.Coal[0],color='black', label='Coal')
plt.bar(3,df.Nuclear[0], color='orange', label='Nuclear')
plt.bar(4,df.Other, color='cyan', label='Other')
plt.bar(5, df.Petroleum[0], color='grey', label='Petroleum')
plt.bar(6,df.Solar,color='yellow', label='Solar')
plt.bar(7, df.Hydroelectric, color='blue', label='Hydroelectric')

plt.xticks(y_pos)
plt.xlabel('Type of Energy')
plt.ylabel('Million Megawatt Hours')
plt.xlim(.25,7.75)
plt.title('Florida Energy Use All Sectors, 2018')
plt.legend()
plt.show()
```
