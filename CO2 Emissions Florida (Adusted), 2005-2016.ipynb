```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel(r'C:\Users\matth\Documents\Excel EIA Energy\CO2 Emissions by state 2005-2017.xlsx')
df1 = df[df.iloc[:]>1].dropna(axis='rows')
df1.index = df1.index + 2003
plt.plot(df1.Florida, color='red')
plt.xlabel('Year')
plt.ylabel('Million Metric Tons CO2')
plt.xlim(2005,2016)
plt.title('CO2 Emissions Florida (Adjusted), 2005-2016')
y_pos = np.arange(2005,2016.25)
plt.xticks(y_pos)

plt.show()
```
