```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel(r'C:\Users\matth\Documents\Excel EIA Energy\CNGRENNUC2.xlsx')
df.index = df.index+9
df = df/1000000

plt.plot(df.Coal, color='black', label='Coal')
plt.plot(df.Natural_Gas, color='red', label='Natural Gas')
plt.plot(df.Nuclear, color='orange', label='Nuclear')
plt.plot(df.Total_Renewable_Generation, color='green', label='Renewable')
plt.ylabel('Billion Megawatt Hours')
plt.xlabel('Year')
plt.title('US Net Energy Generation: \n Coal vs Natural Gas vs Renewable vs. Nuclear \n 2009-2018 (All Sectors)')
plt.xlim(9,18)
plt.legend()
plt.show()
```
